```python
x=date_data.iloc[:, :-1]
y=date_data.iloc[:, -1]
x.head()
y.head()
```

```python
date_data["Class"] = date_data["Class"].astype('category')
date_data.dtypes
```

```python
date_data["Class"] = date_data["Class"].cat.codes
date_data.head()
```

```python
#plot for class values-date fruit dataset has 7 class values as plotted below
import seaborn as sns
sns.countplot(x=date_data['Class'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
#preprocessing
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from importlib import reload
plt=reload(plt)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
neighbors = np.arange(1, 11)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
for i, k in enumerate(neighbors):
    # We instantiate the classifier
    knn = KNeighborsClassifier(n_neighbors=k,p=1)
    # Fit the classifier to the training data
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    print(confusion_matrix(y_test,pred_i))
    print(classification_report(y_test,pred_i))
    #root mean square value:
    #print(sqrt(mean_squared_error( y_test,pred_i)))
    #calculating accuracy for training set
    train_accuracy[i] = knn.score(X_train, y_train)
    #calculating accuracy for training set 
    test_accuracy[i] = knn.score(X_test, y_test)
```

```python
#plot between kvalues and accuracy score:
plt.title('k-NN:Number of neighnors changing')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy',color='blue')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy',color='yellow')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from importlib import reload
plt=reload(plt)
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
neighbors = np.arange(1, 11)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k,p=3)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    print(confusion_matrix(y_test,pred_i))
    print(classification_report(y_test,pred_i))
    #root mean square value:
    #print(sqrt(mean_squared_error( y_test,pred_i)))
    #calculating accuracy for training set
    train_accuracy[i] = knn.score(X_train, y_train)
    #calculating accuracy for testing set
    test_accuracy[i] = knn.score(X_test, y_test)
```

```python
#plot between kvalues and accuracy score:
plt.title('k-NN:Number of neighnors changing')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy',color='blue')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy',color='yellow')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
#applying GridSearch CV for our dataset to find the best parameters in the given set:
knn = KNeighborsClassifier()
from sklearn.model_selection import GridSearchCV
k_range = list(range(1, 35))
param_grid = dict(n_neighbors=k_range)
grid = GridSearchCV(knn, param_grid, cv=10, scoring='accuracy', return_train_score=False,verbose=1)
grid_search=grid.fit(X_train, y_train)
print(grid_search.best_params_)
```

```python
#after getting best K_value:
accuracy = grid_search.best_score_ *100
print("Accuracy for our training dataset with tuning is : {:.2f}%".format(accuracy) )
```

```python
#again applying KNN classifier for best K-value to check accuracy:
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_test_hat=knn.predict(X_test)
test_accuracy=accuracy_score(y_test,y_test_hat)*100
print("Accuracy for our testing dataset with tuning is : {:.2f}%".format(test_accuracy) )
```
